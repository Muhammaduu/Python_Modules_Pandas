Cleaning Empty Rows

dropna() method rows that contain empty cells

```python
df1=df.dropna()
df1
```

Replace Empty Cells Values 

fillna() method replace nan values with an other values instead of deleting row 

Another way of dealing with empty cells is to insert a new value instead.
This way you do not have to delete entire rows just because of some empty cells.
The fillna() method allows us to replace empty cells with a value:

```python
df1.fillna(35)
```

Replacing for only specified columns

To only replace empty values for one column, specify the column name for the DataFrame

```python
df1["CompTotal"]=df1["CompTotal"].fillna(31)
df1["CompTotal"]
```

Replacing NaN values with mean, median and mode 

A common way to replace empty cells, is to calculate the mean, median or mode value of the column.
Pandas uses the mean() median() and mode() methods to calculate the respective values for a specified column

EXAMPLE

```python
# still we are using CompTotal column to demostrate this by fill NaN with mean
a=df1['CompTotal'].mean()# calculating the mean
df1["CompTotal"]=df1["CompTotal"].fillna(a)
df1["CompTotal"]
```

```python
# still we are using CompTotal column to demostrate this by fill NaN with median
a=df1['CompTotal'].median()# calculating the median
df1["CompTotal"]=df1["CompTotal"].fillna(a)
df1["CompTotal"]
```

```python
# still we are using CompTotal column to demostrate this by fill NaN with mode
a=df1['CompTotal'].mode()[0]# calculating the mean
df1["CompTotal"]=df1["CompTotal"].fillna(a)
df1["CompTotal"]
```
